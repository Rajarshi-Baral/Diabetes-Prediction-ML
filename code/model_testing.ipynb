```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Re-import classifiers (if needed)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

# Store results
results = []

for name, clf in models.items():
    clf.fit(X, y)
    y_pred_model = clf.predict(X)
    acc = accuracy_score(y, y_pred_model)
    results.append({"Model": name, "Accuracy": acc})

# Convert results to DataFrame and sort descending
results_df = pd.DataFrame(results).sort_values(by="Accuracy", ascending=False)

# --- Plot Bar Chart ---
plt.figure(figsize=(8, 5))
sns.barplot(
    x="Accuracy", 
    y="Model", 
    data=results_df, 
    palette="viridis",
    hue="Model",
    dodge=False,
    legend=False
)
plt.title("Classifier Accuracy Comparison")
plt.xlabel("Accuracy Score")
plt.ylabel("Classifier")
plt.xlim(0.4, 1.0)
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- Display Accuracy Table ---
print("\nAccuracy Table:\n")
print(results_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import plot_tree

# Create simple 2D classification data for boundary plots
X_vis, y_vis = make_classification(n_samples=300, n_features=2, n_redundant=0,
                                   n_clusters_per_class=1, random_state=42)

# Initialize models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=1, max_depth=3, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=1, max_depth=3, random_state=42),
    'Logistic Regression': LogisticRegression(),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='linear')
}

# Train models
for model in models.values():
    model.fit(X_vis, y_vis)

# Plot setup
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
axes = axes.ravel()

# --- Random Forest Tree ---
plot_tree(models['Random Forest'].estimators_[0], filled=True,
          feature_names=['Feature 1', 'Feature 2'],
          class_names=['Class 0', 'Class 1'], ax=axes[0])
axes[0].set_title('Random Forest (Tree View)')

# --- Gradient Boosting Tree ---
plot_tree(models['Gradient Boosting'].estimators_[0, 0], filled=True,
          feature_names=['Feature 1', 'Feature 2'],
          class_names=['Class 0', 'Class 1'], ax=axes[1])
axes[1].set_title('Gradient Boosting (Tree View)')

# --- Logistic Regression Coefficients as arrows ---
coefs = models['Logistic Regression'].coef_[0]
axes[2].scatter(X_vis[:, 0], X_vis[:, 1], c=y_vis, cmap='coolwarm', edgecolor='k')
axes[2].arrow(0, 0, coefs[0], coefs[1], head_width=0.2, head_length=0.2, fc='black', ec='black')
axes[2].set_title('Logistic Regression (Direction of Decision Boundary)')

# --- KNN Decision Boundary ---
from matplotlib.colors import ListedColormap
h = .02
x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z_knn = models['KNN'].predict(np.c_[xx.ravel(), yy.ravel()])
Z_knn = Z_knn.reshape(xx.shape)
axes[3].contourf(xx, yy, Z_knn, alpha=0.4, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
axes[3].scatter(X_vis[:, 0], X_vis[:, 1], c=y_vis, cmap='coolwarm', edgecolor='k')
axes[3].set_title('K-Nearest Neighbors (Decision Boundary)')

# --- SVM Decision Boundary ---
Z_svm = models['SVM'].predict(np.c_[xx.ravel(), yy.ravel()])
Z_svm = Z_svm.reshape(xx.shape)
axes[4].contourf(xx, yy, Z_svm, alpha=0.4, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
axes[4].scatter(X_vis[:, 0], X_vis[:, 1], c=y_vis, cmap='coolwarm', edgecolor='k')
axes[4].set_title('Support Vector Machine (Decision Boundary)')

# Hide empty subplot
axes[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load dataset
df = pd.read_csv('../dataset/diabetes_cleaned.csv')

# Prepare features and labels
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Select only first two features for 2D plotting
X_vis = X.iloc[:, :2]
y_vis = y

# Standardize features for decision boundary
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_vis)

# Define classifiers
models = {
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
}

# --- Plot Decision Boundaries ---
plt.figure(figsize=(12, 10))
xx, yy = np.meshgrid(np.linspace(X_scaled[:, 0].min()-1, X_scaled[:, 0].max()+1, 500),
                     np.linspace(X_scaled[:, 1].min()-1, X_scaled[:, 1].max()+1, 500))

for i, (name, model) in enumerate(models.items()):
    model.fit(X_scaled, y_vis)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(2, 2, i+1)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_vis, palette="coolwarm", s=30, edgecolor="k", legend=False)
    plt.title(f"{name} Decision Boundary")
    plt.xlabel(X_vis.columns[0])
    plt.ylabel(X_vis.columns[1])

plt.suptitle("Classifier Decision Boundaries (Using First 2 Features)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# --- Plot Decision Tree ---
plt.figure(figsize=(15, 8))
tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X, y)
plot_tree(tree_clf, feature_names=X.columns, class_names=["Not Diabetic", "Diabetic"], filled=True, rounded=True)
plt.title("Decision Tree Classifier (max_depth=3)")
plt.show()
```
